```python
# se limita la cantidad de registros del dataframe
cant_registros = 5403
df = df.iloc[0:5403]

# elimina columnas innecesarias
cols_drop = ['id', 'masl', 'systolic_bp', 'diastolic_bp', 'body_mass_index', 'dm_treatment', 'cd_treatment', 'smoking', 
             'hypertension_dx', 'hypertension_years', 'hypertension_treatment', 'sist_old', 'diast_old', 'sist_new', 
             'diast_new', 'treatment ']

#invoices.drop(['invoice', 'client', 'units'], axis=1)

df.drop(cols_drop, axis = 'columns', inplace=True)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=["diabetes_mellitus", "cv_diseases"])
df.isna().sum()
```

```python
# correccion de nombres de columna y valores

df.rename(columns={'treatment ': 'treatment'}, inplace=True)

df['BMI_cat'] = df['BMI_cat'].replace(['obese'],['Obese'])
# df.columns
```

```python
# verifica las variables usadas dentro de cada columna agrupándolas
cols = df.columns
for i in cols:
    print(df.groupby(i).mean())
```

```python
# convertir variables categoricas a numericas
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df['city'] = encoder.fit_transform(df.city.values)
df['sex'] = encoder.fit_transform(df.sex.values)
df['diabetes_mellitus'] = encoder.fit_transform(df.diabetes_mellitus.values)
#df['dm_treatment'] = encoder.fit_transform(df.dm_treatment.values)
df['cv_diseases'] = encoder.fit_transform(df.cv_diseases.values)
#df['cd_treatment'] = encoder.fit_transform(df.cd_treatment.values)
#df['smoking'] = encoder.fit_transform(df.smoking.values)
#df['hypertension_dx'] = encoder.fit_transform(df.hypertension_dx.values)
#df['hypertension_treatment'] = encoder.fit_transform(df.hypertension_treatment.values)
df['physical_activity'] = encoder.fit_transform(df.physical_activity.values)
df['msnm'] = encoder.fit_transform(df.msnm.values) # ordenarlos??
df['region'] = encoder.fit_transform(df.region.values)
df['msnm'] = encoder.fit_transform(df.msnm.values)
#df['sist_old'] = encoder.fit_transform(df.sist_old.values)
#df['diast_old'] = encoder.fit_transform(df.diast_old.values)
#df['sist_new'] = encoder.fit_transform(df.sist_new.values)
#df['diast_new'] = encoder.fit_transform(df.diast_new.values)
#df['treatment'] = encoder.fit_transform(df.treatment.values)
df['HTA_new'] = encoder.fit_transform(df.HTA_new.values)
df['BMI_cat'] = encoder.fit_transform(df.BMI_cat.values)

df
```
