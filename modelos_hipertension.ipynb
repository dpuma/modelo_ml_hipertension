```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
path = 'Change criteria hypertension peru.xls'
sheet_name='Sheet1'

df = pd.read_excel(path, sheet_name)
df
```

```python
# se limita la cantidad de registros del dataframe
cant_registros = 5403
df = df.iloc[0:5403]

# se eliminan las columnas innecesarias
cols_drop  = ['id', 'cd_treatment',  'masl',  'body_mass_index',  'smoking', 
            'hypertension_years','dm_treatment', 'hypertension_dx', 'hypertension_treatment', 'msnm', 
            'sist_old', 'diast_old', 'sist_new', 'diast_new','treatment ','BMI_cat','systolic_bp','diastolic_bp']

df.drop(cols_drop, axis = 'columns', inplace=True)
```

```python
df.isna().sum()
```

```python
# se convierten los datos nan de la columna smoking_years a cero
df['smoking_years'] = df['smoking_years'].replace(np.nan, 0)

# se eliminan los registros con datos ausentes(nan) de las otras columnas 
df = df.dropna(subset=["diabetes_mellitus", "cv_diseases"])

# se reestablece el orden de los indices
df.reset_index(drop=True, inplace=True)
```

```python
# histogramas
cols = df.columns.values
fig = plt.figure(1, (20,16))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(0, len(cols)):
    plt.subplot(4, 5, i+1)
    plt.hist(df[cols[i]])
    plt.title(cols[i])
```

```python
# convertir variables categoricas a numericas
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df['city'] = encoder.fit_transform(df.city.values)
df['sex'] = encoder.fit_transform(df.sex.values)
df['diabetes_mellitus'] = encoder.fit_transform(df.diabetes_mellitus.values)
df['cv_diseases'] = encoder.fit_transform(df.cv_diseases.values)
df['physical_activity'] = encoder.fit_transform(df.physical_activity.values)
df['region'] = encoder.fit_transform(df.region.values)
df['HTA_new'] = encoder.fit_transform(df.HTA_new.values)

df
```

```python
df.info()
```

```python
# mapa de calor
plt.figure(figsize=(14, 8))

sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# split de grupos de entrenamiento y prueba
X = np.array(df.drop(columns=['HTA_new']))
y = df['HTA_new']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=3)
```

```python
model_rf = RandomForestClassifier(n_estimators = 500, random_state = 100)
model_rf.fit(X_train, y_train)
```

```python
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

y_pred = model_rf.predict(X_test)
matriz_rf = confusion_matrix(y_test, y_pred)

plot_confusion_matrix(conf_mat = matriz_rf, figsize=(6,6), show_normed = False)
plt.tight_layout()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
#rom sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
#from sklearn.model_selection import KFold

# cross validation
accuracy_rf  = cross_val_score(model_rf,X_test,y_test, cv=10, scoring='accuracy').mean()
precision_rf = cross_val_score(model_rf,X_test,y_test, cv=10, scoring='precision_macro').mean() 
recall_rf    = cross_val_score(model_rf,X_test,y_test, cv=10, scoring='recall_macro').mean()
f1_rf        = cross_val_score(model_rf,X_test,y_test, cv=10, scoring='f1_macro').mean()
print('Accuracy  :', accuracy_rf) 
print('Precision :', precision_rf) 
print('Recall    :', recall_rf) 
print('f1-score  :', f1_rf)
```

```python
import lightgbm as lgb

model_lgbm = lgb.LGBMClassifier(n_estimators = 500, random_state = 100)
model_lgbm.fit(X_train, y_train)
```

```python
# predict the results
y_pred = model_lgbm.predict(X_test)
matriz_lgbm = confusion_matrix(y_test, y_pred)

plot_confusion_matrix(conf_mat = matriz_lgbm, figsize=(6,6), show_normed = False)
plt.tight_layout()
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# cross validation
accuracy_lgbm  = cross_val_score(model_lgbm,X_test,y_test, cv=10, scoring='accuracy').mean()
precision_lgbm = cross_val_score(model_lgbm,X_test,y_test, cv=10, scoring='precision_macro').mean() 
recall_lgbm    = cross_val_score(model_lgbm,X_test,y_test, cv=10, scoring='recall_macro').mean()
f1_lgbm        = cross_val_score(model_lgbm,X_test,y_test, cv=10, scoring='f1_macro').mean()
print('Accuracy  :', accuracy_lgbm) 
print('Precision :', precision_lgbm) 
print('Recall    :', recall_lgbm) 
print('f1-score  :', f1_lgbm)
```

```python
# Ploteo de mÃ©tricas de evaluaciÃ³n
datos = {
    'RF' : [accuracy_rf, precision_rf, recall_rf, f1_rf],
    'LightGBM': [accuracy_lgbm, precision_lgbm, recall_lgbm, f1_lgbm],
}
resultados = pd.DataFrame(datos)
resultados.index = ['Accuracy', 'Precision', 'Recall', 'f1-score']
print(resultados)
resultados.plot(kind='bar')
plt.ylim(0,1)
plt.title("MÃ©tricas de validaciÃ³n de los modelos")
plt.xlabel("MÃ©trica")
plt.ylabel("Valor")

plt.show()
```
